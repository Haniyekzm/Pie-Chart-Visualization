```python
import pandas as pd
import matplotlib.pyplot as plt

# تعریف مسیر فایل اکسل
file_path = 'full_preprocessed_5H2O2_run2.csv'  # مسیر صحیح فایل اکسل خود را بگذارید

# خواندن دادهها از فایل اکسل
df = pd.read_csv(file_path)

# انتخاب ستونهای Residue و TOTAL
data = df[['Residue', 'TOTAL']]

# جدا کردن residue هایی با TOTAL مثبت و صفر
positive_data = data[data['TOTAL'] >= 0]

# جدا کردن residue هایی با TOTAL منفی
negative_data = data[data['TOTAL'] < 0]

# محاسبه تعداد residue ها و درصدها برای residue هایی که TOTAL مثبت و صفر دارند
positive_counts = positive_data['Residue'].value_counts()

# محاسبه درصد residue های مثبت
if not positive_counts.empty:
    plt.figure(figsize=(10, 6))
    plt.pie(positive_counts, labels=positive_counts.index, autopct='%1.1f%%', startangle=90)
    plt.axis('equal')  # برای اینکه دایره واقعی باشد
    plt.title('Percentage of Positive or Zero Residues by Total')

    # ذخیره نمودار به عنوان یک تصویر
    plt.savefig('positive_residues_pie_chart2.png')
    plt.show()
else:
    print("هیچ residue مثبت یا صفر برای رسم نمودار وجود ندارد.")

# محاسبه تعداد residue ها و درصدها برای residue هایی که TOTAL منفی دارند
negative_counts = negative_data['Residue'].value_counts()

# محاسبه درصد residue های منفی
if not negative_counts.empty:
    plt.figure(figsize=(10, 6))
    plt.pie(negative_counts, labels=negative_counts.index, autopct='%1.1f%%', startangle=90)
    plt.axis('equal')  # برای اینکه دایره واقعی باشد
    plt.title('Percentage of Negative Residues by Total')

    # ذخیره نمودار به عنوان یک تصویر
    plt.savefig('negative_residues_pie_chart2.png')
    plt.show()
else:
    print("هیچ residue منفی برای رسم نمودار وجود ندارد.")
```
