```python
import pandas as pd
import matplotlib.pyplot as plt

# خواندن دادهها از فایل اکسل
file_path = 'full_preprocessed_5H2O2_run1.csv'  # مسیر فایل اکسل خود را در اینجا قرار دهید
df = pd.read_csv(file_path)

# فرض کنید ستونها به نام residue و total هستند
# مرتبسازی بر اساس total از منفی به بزرگ
sorted_df = df.sort_values(by='TOTAL', ascending=True)

# محاسبه تعداد residue هایی که total منفی دارند
negative_total = sorted_df[sorted_df['TOTAL'] < 0]
positive_total = sorted_df[sorted_df['TOTAL'] >= 0]

# محاسبه درصد residue ها
negative_count = len(negative_total)
positive_count = len(positive_total)
total_count = negative_count + positive_count

negative_percentage = (negative_count*100 / total_count)
positive_percentage = (positive_count *100/ total_count)

# ذخیره مقادیر در یک DataFrame برای خروجی
output_df = pd.DataFrame({
    'Category': ['Negative Total', 'Positive Total', 'Negative Percentage'],
    'Count': [negative_count, positive_count, negative_percentage]
})

# ذخیره خروجی به فایل CSV
output_df.to_csv('output_counts.csv', index=False)

# رسم pie chart برای residue هایی با total منفی
plt.figure(figsize=(15, 5))

# Pie chart برای residue هایی با total منفی
plt.subplot(1, 3, 1)
plt.pie([negative_count, total_count - negative_count], 
        labels=['Negative Total', 'Not Negative Total'], 
        autopct='%1.1f%%', 
        startangle=90)
plt.title('Residues with Negative Total')

# Pie chart برای residue هایی با total مثبت
plt.subplot(1, 3, 2)
plt.pie([positive_count, total_count - positive_count], 
        labels=['Positive Total', 'Not Positive Total'], 
        autopct='%1.1f%%', 
        startangle=90)
plt.title('Residues with Positive Total')

# Pie chart برای درصد همه residue ها با total منفی
plt.subplot(1, 3, 3)
plt.pie([negative_percentage, 100 - negative_percentage], 
        labels=['Negative Total Percentage', 'Positive Total Percentage'], 
        autopct='%1.1f%%', 
        startangle=90)
plt.title('Percentage of Residues with Negative Total')

plt.tight_layout()
plt.savefig('output_charts.png')  # ذخیره نمودارها به عنوان فایل PNG
plt.show()
```
